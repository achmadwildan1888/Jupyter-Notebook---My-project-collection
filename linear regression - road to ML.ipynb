# Scikit learn


```python
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn')
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])

df= pd.DataFrame({
    'x':x,
    'y':y
})
df
```

```python
plt.plot(df['x'],df['y'],'r')
plt.scatter(df['x'],df['y'],color ='r',marker='o',s=50)
plt.xlabel('nilai x')
plt.ylabel('nilai y')
plt.title('plot awal')
plt.show()
```

```python
#linear rreg

from sklearn import linear_model

model = linear_model.LinearRegression()

#training dataset :model.fit(dataindependend , data dependent)
model.fit(df[['x']],df['y']) #data independednt harus 2D , dependent 1D


#show m(gradient/slope)  & b(intercept) of 'best fit line'
print('slope = ', model.coef_)
print('intercept = ', model.intercept_)

print(model.predict([[8]]))



```

```python
plt.plot(df['x'],df['y'],'r')
plt.scatter(df['x'],df['y'],color ='r',marker='o',s=50)

#memasukan hasil prediksi ke dataframe
plt.plot(df['x'],model.predict(df[['x']]),'g')
plt.scatter(df['x'],model.predict(df[['x']]),color ='g',marker='o',s=50)
plt.xlabel('nilai x')
plt.ylabel('nilai prediksi')
plt.title('plot prediksi')
plt.show()
```

```python
tanah = pd.read_csv('prediksitanah.csv')
tanah

model = linear_model.LinearRegression()

#training dataset :model.fit(dataindependend , data dependent)
model.fit(tanah[['luas']],tanah['harga']) #data independednt harus 2D , dependent 1D


#show m(gradient/slope)  & b(intercept) of 'best fit line'
print('slope = ', model.coef_)
print('intercept = ', model.intercept_)

print(model.predict([[5000]]))
```

```python
plt.plot(tanah['luas'],tanah['harga'],'r')
plt.scatter(tanah['luas'],tanah['harga'],color ='r',marker='o',s=50)
plt.xlabel('nilai luas')
plt.ylabel('nilai harga')
plt.title('plot prediksi harga tanah')
plt.show()
```

```python
plt.plot(tanah['luas'],tanah['harga'],'r')
plt.scatter(tanah['luas'],tanah['harga'],color ='r',marker='o',s=50)

#memasukan hasil prediksi ke dataframe
plt.plot(tanah['luas'],model.predict(tanah[['luas']]),'g')
plt.scatter(tanah['luas'],model.predict(tanah[['luas']]),color ='g',marker='o',s=50)
plt.xlabel('nilai luas')
plt.ylabel('nilai harga')
plt.title('plot prediksi harga tanah')
#accuracy
print('akurasinya= ',model.score(tanah[['luas']],tanah['harga']))
plt.show()
```
