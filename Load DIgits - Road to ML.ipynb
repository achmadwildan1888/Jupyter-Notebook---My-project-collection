```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.datasets import load_digits
from sklearn import linear_model
import seaborn as sns
```

```python
digits = load_digits()
```

```python
print(digits['target'][13])
fig = plt.figure('Digit', figsize = (6,6))
plt.imshow(digits['images'][13],cmap = 'gray_r')
plt.show
```

```python
fig = plt.figure('digit',figsize=(9,3))
plt.subplot(131)
plt.imshow(digits['images'][0],cmap='gray')
plt.title('angka={}'.format(digits['target'][0]))
plt.subplot(132)
plt.imshow(digits['images'][1],cmap='gray')
plt.title('angka={}'.format(digits['target'][1]))
plt.subplot(133)
plt.imshow(digits['images'][2],cmap='gray')
plt.title('angka={}'.format(digits['target'][2]))                            
plt.show()
```

```python
#atau pake for loop
fig = plt.figure('digit',figsize=(12,6)) #baris kolom (5 baris 2 kolom)

for x in range(10):
    plt.subplot(2,5,x+1) #x+1 yaitu untuk posisi
    plt.imshow(digits['images'][x],cmap='gray')
    plt.title('angka={}'.format(digits['target'][x]))
plt.show()
```

```python
user= (input('ketik tanggal lahir ='))

fig = plt.figure('digit',figsize=(12,2)) #figure 6 digit (12 dibagi 2)
count=1
for x in user:
    plt.subplot(1,6,count) #x+1 yaitu untuk posisi
    plt.imshow(digits['images'][int(x)],cmap='gray')
    plt.title('angka={}'.format(digits['target'][int(x)]))
    count+=1
plt.show()
```
