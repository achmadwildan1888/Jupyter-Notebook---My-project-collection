```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data_tree.csv')
df.head()
```

```python
#convert nominal jadi ordinal
from sklearn.preprocessing import LabelEncoder
labelkantor = LabelEncoder()
labeljabatan = LabelEncoder()
labeltitel = LabelEncoder()

df['kantorLE'] = labelkantor.fit_transform(df['kantor']) #bikin dataframe kantorLE dijadikan ordinal
df['jabatLE'] = labelkantor.fit_transform(df['jabatan']) 
df['titelLE'] = labelkantor.fit_transform(df['titel']) 
```

```python
df
```

```python
df = df.drop(
    ['kantor','jabatan','titel'],axis=1
)
    
```

```python
df 

#kantor = 0 ,facebook, 1 google , 2 tesla
#jabatan : 0 GM , 1 Manager,  2 Staff
#Titel : 0 s1 , 1 s2
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()

#train model
model.fit(df[['kantorLE','jabatLE','titelLE']],df['gaji>50'])
```

```python

#train model
acc = model.score(df[['kantorLE','jabatLE','titelLE']],df['gaji>50'])
acc
```

```python
#predict kantor, jabatan , titel
print(model.predict([[0,0,0]]))
print(model.predict([[2,0,0]]))
```

```python
#coba dengan split train 80 %
from sklearn.model_selection import train_test_split
x_train,x_tes,y_train,y_tes = train_test_split(df[['kantorLE','jabatLE','titelLE']],df['gaji>50'],
                test_size=.2,random_state = 1)
```

```python
print(len(x_train))
print(len(y_tes))
```

```python

#train model
model.fit(x_train,y_train)
akurasi = model.score(x_train,y_train)
akurasi
```
