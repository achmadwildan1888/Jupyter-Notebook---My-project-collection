```python
import sklearn as sk
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
plt.style.use('ggplot')
%matplotlib notebook
```

```python

data = {
    'luas' : [2600,3000,3200,3600,4000],
    'kamar':[3,4,1,3,5],
    'harga' : [550000,565000,610000,595000,760000]
}

df= pd.DataFrame(data)
df
```

```python
model = linear_model.LinearRegression()
#training dataset :model.fit(dataindependend , data dependent)
model.fit(df[['luas','kamar']],df['harga']) #data independednt harus 2D , dependent 1D
```

```python
#m1x1 + m2x2 +b
#slope jadi 2 
print('slope = ', model.coef_)
print('intercept = ', model.intercept_)
```

```python
#prediksi
print(model.predict([[3400,3]]))
```

```python
#plotting luas & harga
plt.figure('multivariate regress' , figsize=(6,3)) #FIGURE 3X3 DAN 3X3

#plot luas &harga
plt.subplot(1,2,1)
plt.plot(df['luas'],df['harga'], 'r-')


plt.subplot(1,2,2)
plt.plot(df['kamar'],df['harga'],'b-')

plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure('multivarete reg')
ax=plt.subplot(111,projection = '3d')

ax.plot_wireframe(df['luas'],df['kamar'],df[['harga']])

ax.set_xlabel('luas tanah')
ax.set_ylabel('n kamar')
ax.set_zlabel('harga rmh')
```

```python
#coba 3 variable
data = {
    'luas' : [2600,3000,3200,3600,4000],
    'kamar':[3,4,1,3,5],
    'usia' : [20,15,18,30,8],
    'harga' : [550000,565000,610000,595000,760000]
}

#training

df= pd.DataFrame(data)
```

```python

model = linear_model.LinearRegression()
#training dataset :model.fit(dataindependend , data dependent)
model.fit(df[['luas','kamar','usia']],df['harga']) #data independednt harus 2D , dependent 1D

#m1x1 + m2x2 +b
#slope jadi 2 
print('slope = ', model.coef_)
print('intercept = ', model.intercept_)
```

```python
#prediksi
print(model.predict([[3200,2,10]]))
```

```python
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure('multivarete reg 3 var')
ax=plt.subplot(111,projection = '3d')

plot = ax.scatter(
           df['luas'],
           df['kamar'],
           df[['harga']],
           c=df.usia, 
           marker='o',
           s=120,
           cmap='hot_r'
)

#harga terbaik
ax.scatter(
           df['luas'],
           df['kamar'],
           model.predict(df[['luas','kamar','usia']]),
           marker='*',
           s=175,
           color='green'
)

fig.colorbar(plot)
ax.set_xlabel('luas tanah')
ax.set_ylabel('n kamar')
ax.set_zlabel('harga rmh')
```
