```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
```

```python
wajah = fetch_olivetti_faces()
```

```python
dir(wajah)
print(len(wajah['data'][0])) # 64 x 64
```

```python
fig = plt.figure('wajah',figsize=(14,7))
for i in range (10): #ada 10 pose
    person=11
    plt.subplot(2,5,i+1)  #2 baris , 5 kolom, pose ke i+1 karena subplot dimulai dr 1
    plt.imshow(wajah['images'][i +(10*person)],cmap = 'gray')
    plt.suptitle('wajah orang ke {}'.format(person))
```

```python
#split datasets : train 90% , test = 10%

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
            wajah['data'],
            wajah['target'],
            test_size = .1)
```

```python
#support vector machine
#support vector classifier

from sklearn.svm import SVC
model = SVC(gamma = 'auto')

model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
print(model.predict([x_test[0]]))
print(y_test[0])
```
