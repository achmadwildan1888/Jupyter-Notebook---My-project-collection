```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.datasets import load_boston
from sklearn import linear_model
```

```python
databoston=load_boston()
```

```python
dir(databoston)
```

```python
databoston['data'].shape
```

```python
databoston['feature_names']
```

```python
databoston
```

```python
boston = pd.DataFrame(databoston['data'],columns=databoston['feature_names'])
print(boston)
```

```python
#buat median value
boston['MEDV'] = databoston['target']
boston
```

```python
model = linear_model.LinearRegression()

#training
model.fit(boston[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD','TAX', 'PTRATIO', 'B', 'LSTAT']], boston['MEDV'])
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
#akurasinya
print(model.score(boston[
    ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD','TAX', 'PTRATIO', 'B', 'LSTAT']
    ], boston['MEDV']))
```

```python
#memprediksi data 1 apakah 24
print(model.predict([[
    0.00632,18.0,2.31,0.0,0.538,6.575,65.2,4.0900,1.0,296.0,15.3,396.90,4.98]
]))
```
