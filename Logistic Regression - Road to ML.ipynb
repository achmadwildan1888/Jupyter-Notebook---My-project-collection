```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk
from sklearn.model_selection import train_test_split 
```

```python
df = pd.read_csv('data1.csv')
```

```python
plt.figure(figsize=(12,5))
plt.scatter(x=df.usia,y=df.beliasuransi)
plt.title('data pembeli asuransi')
plt.xlabel('usia')
plt.ylabel('beli atau tidak',rotation='horizontal')
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[['usia']],df.beliasuransi,test_size = .1)
```

```python
#logistic regression

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs')
```

```python
#training model
model.fit(x_train,y_train)
```

```python
#slope coeff n intercept
print(model.coef_)
print(model.intercept_)
```

```python
print(x_test)
print(model.predict(x_test)) #dari xtest akankah beli atau tidak
```

```python
print(model.score(x_train,y_train))
print(model.predict_proba([[2]])) #95% tidak akan beli asuransi untuk umur 2 tahun
```

```python
plt.figure(figsize=(12,5))
plt.scatter(x=df.usia,y=df.beliasuransi)


plt.title('data pembeli asuransi')
plt.xlabel('usia')
plt.ylabel('beli atau tidak',rotation='horizontal')


dataplot = np.linspace(0,70,150)
def plotlogreg(line):
    return 1/(1+ np.exp(-line))

bestfit = plotlogreg((model.coef_ *dataplot) + model.intercept_).ravel()


# rumusnya
# 1 / (1+ exp( mx + b))
#plot best fitline
plt.plot(
        dataplot,
        bestfit,
        
)
```
