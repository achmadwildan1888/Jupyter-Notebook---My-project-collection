```python
import pandas as pd

dftelkom = pd.read_csv(
    'TLKM.JK.csv',
    parse_dates=['Date'], #merubah kolom date jadi type data tanggal , jika tidak diubah menjadi date tdk bisa dipanggil
    index_col='Date')

dfXL = pd.read_csv(
    'EXCL.JK.csv',
    parse_dates=['Date'], #merubah kolom date jadi type data tanggal , jika tidak diubah menjadi date tdk bisa dipanggil
    index_col='Date')

dfindosat = pd.read_csv(
    'ISAT.JK.csv',
    parse_dates=['Date'], #merubah kolom date jadi type data tanggal , jika tidak diubah menjadi date tdk bisa dipanggil
    index_col='Date')

dfsmart = pd.read_csv(
    'FREN.JK.csv',
    parse_dates=['Date'], #merubah kolom date jadi type data tanggal , jika tidak diubah menjadi date tdk bisa dipanggil
    index_col='Date')
```

```python
dftelkom = dftelkom ['Close'].resample('W').max().fillna(
    method='ffill'
)

dfXL = dfXL ['Close'].resample('W').max().fillna(
    method='ffill'
)
dfindosat = dfindosat ['Close'].resample('W').max().fillna(
    method='ffill'
)
dfsmart = dfsmart ['Close'].resample('W').max().fillna(
    method='ffill'
)
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')

plt.plot(dftelkom,'r-')
plt.plot(dfXL,'b-')
plt.plot(dfindosat,'y-')
plt.plot(dfsmart,'m-')

plt.plot(dftelkom,'ro')
plt.plot(dfXL,'bo')
plt.plot(dfindosat,'yo')
plt.plot(dfsmart,'mo')

plt.legend(['Telkom','XL','indosat','smartfren'])
plt.grid(True)

plt.xlabel('Tanggal')
plt.ylabel('Close')

plt.show()
```
